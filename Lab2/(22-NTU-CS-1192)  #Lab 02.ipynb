```python
import pandas as pd
df2=pd.read_csv("Data Set/headbrain.csv")
```

```python
df2.head()
```

```python
X=df2['Head Size(cm^3)'].values
y=df2['Brain Weight(grams)'].values
```

```python
import numpy as np
X_sum=np.sum(X)
X_squared_sum=np.sum(X*X)
n=len(X)
print(f"Sum of X: {X_sum}")
print(f"Sum of X Squared: {X_squared_sum}")
print(f"Length of X: {n}")
```

```python
numerator=n*np.sum(X*y)-np.sum(X)*np.sum(y)
denominator=n*np.sum(X*X)-(np.sum(X))**2
w1=numerator/denominator
w0=(np.sum(y)-w1*(np.sum(X)))/n

print(f"w0: {w0}")
print(f"w1: {w1}")
```

```python
max_x=np.max(X)
min_x=np.min(X)

x1=np.linspace(min_x,max_x)
y1=w0+w1*x1

plt.plot(x1,y1 ,color='red',alpha=0.9, label='Regression Line')
plt.scatter(X, y, c='seagreen', label='Scatter Plot')
plt.xlabel("Head Size (cm^3)")
plt.ylabel("Brain Weight (grams)")
plt.legend()
plt.grid(True)
plt.show()

         
```

```python
rmse=0
for i in range(n):
    y_pred =w0 + w1 * X[i]
    rmse+= (y[i] - y_pred) ** 2
rmse=np.sqrt(rmse/n)
print("RMSE=",rmse)
```

```python
ss_tot=0
ss_res=0
y_mean=np.mean(y)
for i in range(n):
    y_pred=w0+w1*X[i]
    ss_tot+=(y[i]-y_mean)**2
    ss_res+=(y[i]-y_pred)**2
r2=1-(ss_res/ss_tot)
print("R2Score=",r2)
```

# 2. Preprocess:


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler


num_cols = ["age", "bmi", "children"]
cat_cols = ["sex", "smoker", "region"]
```

```python
# scale numerical
scaler = StandardScaler()
scaled = scaler.fit_transform(df[num_cols])
```

```python
# encode categorical
encoder = OneHotEncoder(drop="first", sparse_output=False)
encoded = encoder.fit_transform(df[cat_cols])
```

```python
# convert to DataFrames with correct column names & same index
scaled_df = pd.DataFrame(scaled, index=df.index, columns=num_cols)
encoded_df = pd.DataFrame(encoded, index=df.index,
                          columns=encoder.get_feature_names_out(cat_cols))
```

```python
# concat together
X_processed = pd.concat([scaled_df, encoded_df], axis=1)

# target
y = df["charges"]

print(X_processed.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train,X_test,y_train,y_test=train_test_split(
    X_processed,
    y,
    test_size=0.2,
    random_state=0
)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred[:5]
```

```python
data_new=X_train[109:110]
lr.predict(data_new)
```

```python
y_train[109:110]
```

```python
lr.score(X_train,y_train)*100
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,y_pred,c='blue', alpha=0.6, label='Predicted vs Aatual')
plt.plot([y_test.min(), y_test.max()], 
         [y_test.min(), y_test.max()], 
         color='red', linestyle='--', label='Ideal Fit')

plt.title("Actual vs Predicted Charges")
plt.xlabel("Acutal Charges")
plt.ylabel("Predicted Charges")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error

print(f"R 2 Squared: {r2_score(y_test,y_pred):.2f}")
print(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test,y_pred)):.2f}")
```
